## Question 4

```
using System;
public class EvenOdd
{
    public static void Main(string[] args)
    {
        int[] arr = new int[5];
        arr[0] = 25;
        arr[1] = 47;
        arr[2] = 42;
        arr[3] = 56;
        arr[4] = 32;
        int[] arr2 = new int[5];
        int[] arr3 = new int[5];
        int i, j = 0, k = 0;
        for (i = 0; i < 5; i++)
        {
            if (arr[i] % 2 == 0)
            {
                arr2[j] = arr[i];
                j++;
            }
            else
            {
                arr3[k] = arr[i];
                k++;
            }
        }
        Console.WriteLine("The Even elements are:");
        for (i = 0; i < j; i++)
        {
            Console.Write(arr2[i] + "\t");
        }
        Console.WriteLine("\n");
        Console.WriteLine("The Odd elements are:");
        for (i = 0; i < k; i++)
        {
            Console.Write(arr3[i] + "\t");
        }
    }
}
```
